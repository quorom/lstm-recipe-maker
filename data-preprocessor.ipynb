```python
import json
from pprint import pprint
from collections import defaultdict
import string
```

```python
# Idea create recipe maker
```

```python
data = None
with open('recipes.json') as f:
    data = json.load(f)
```

```python
def createRecipeTxt(title, ingred, instr):
    title_txt = "Title: " + title + '\n'
    ingred_txt = "\nIngredients:\n"
    for i in ingred:
        i = ' '.join(i.replace('ADVERTISEMENT', '').split())
        if i != '':
            ingred_txt += " - " + i + '\n'
    
    instr_txt = "\nInstructions:\n"
    instr_txt += instr
    
    recipe = title_txt + ingred_txt + instr_txt + '\n'
    recipe = ''.join(list(filter(lambda x: x in string.printable, recipe)))
    return recipe
```

```python
cookbook = ""
for recipe in data:
    try:
        title = data[recipe]['title']
        ingred = data[recipe]['ingredients']
        instrs = data[recipe]['instructions']
        cookbook += createRecipeTxt(title,ingred,instrs)
    except:
        pass
```
